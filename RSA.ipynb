```python
p=11
q=17
e=3
N=p*q
phi=(p-1)*(q-1)

for i in range(phi):
  if (e*i)%phi==1:
    d=i
    break
print(d)

m=3
c=(m**e)%N
print(c)
md=(c**d)%N
print(md)
```

```python
import string

def msg2int(msg):

  table={" ": 0 }
  for i,s in enumerate(string.ascii_lowercase,start=1):
    table[s]=i
  for i,s in enumerate(string.ascii_uppercase,start=1):
    table[s]=i

  integer=0
  for s in msg:
    integer=integer*100+table[s]
  return integer

msg="Thank you for solving me"
msg2int(msg)
```
