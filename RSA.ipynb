```python
# RSA small example

p=11
q=17
e=3
N=p*q
phi=(p-1)*(q-1)

for i in range(phi):
  if (e*i)%phi==1:
    d=i
    break

print("pk: N=",N,", e=",e)
print("sk: p=",p,",q=",q,",d=",d)

m=3
print("m=",m)
c=(m**e)%N
print("c=",c)
mprime=(c**d)%N
print("m'=",mprime)
```

```python
import string

def msg2int(msg):

  table={" ": 0 }
  for i,s in enumerate(string.ascii_lowercase,start=1):
    table[s]=i
  for i,s in enumerate(string.ascii_uppercase,start=1):
    table[s]=i

  integer=0
  for s in msg:
    integer=integer*100+table[s]
  return integer

def int2msg(integer):
  table={0: " "}
  for i,s in enumerate(string.ascii_lowercase,start=1):
    table[i]=s
  #for i,s in enumerate(string.ascii_uppercase,start=1):
  #  table[i]=s

  msg=""
  while integer>0:
    msg=table[integer%100]+msg
    integer=integer//100
  return msg

msg="Thank you for solving me"
m_int=msg2int(msg)
print(m_int)
print(int2msg(m_int))

import math
print(math.log10(m_int))
#print(math.sqrt(m_int))
```
