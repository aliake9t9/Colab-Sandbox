```python
p=11
q=17
e=3
N=p*q
phi=(p-1)*(q-1)

for i in range(phi):
  if (e*i)%phi==1:
    d=i
    break
print(d)

m=3
c=(m**e)%N
print(c)
md=(c**d)%N
print(md)
```
