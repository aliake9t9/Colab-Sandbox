```python
# RSA small example

p=11
q=17
e=3
N=p*q
phi=(p-1)*(q-1)

def ext_gcd(a, b):
    if b == 0:
        return 1, 0, a
    x1, y1, gcd = ext_gcd(b, a % b)
    x = y1
    y = x1 - (a // b) * y1
    return x, y, gcd

d,_,gcd=ext_gcd(e,phi)
if gcd!=1:
  print("Error")
  exit()

d=d%phi

print("pk: N=",N,", e=",e)
print("sk: p=",p,",q=",q,",d=",d)

m=3
print("m=",m)
c=pow(m,e,N)
print("c=",c)
mprime=pow(c,d,N)
print("m'=",mprime)
```

```python
import string

def msg2int(msg):

  table={" ": 0 }
  for i,s in enumerate(string.ascii_lowercase,start=1):
    table[s]=i
  for i,s in enumerate(string.ascii_uppercase,start=1):
    table[s]=i

  integer=0
  for s in msg:
    integer=integer*100+table[s]
  return integer

def int2msg(integer):
  table={0: " "}
  for i,s in enumerate(string.ascii_lowercase,start=1):
    table[i]=s
  #for i,s in enumerate(string.ascii_uppercase,start=1):
  #  table[i]=s

  msg=""
  while integer>0:
    msg=table[integer%100]+msg
    integer=integer//100
  return msg

msg="Thank you for solving me"
#msg="The quick brown fox jumps over the lazy dog"
m_int=msg2int(msg)
print(m_int)
print(int2msg(m_int))

import math
print(math.log10(m_int))
#print(math.sqrt(m_int))
```
